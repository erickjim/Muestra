# Proceso de Encuesta "Experiencia Empleado"

```python
import pandas as pd
```

```python
# Inserta ruta y nombre del archivo 'Experiencia de Empleado' correspondiente al periodo.

ex_emp='Experiencia de Empleado_8 noviembre 2021_09.59.csv'
```

```python
# Lectura de Universo Telcel y Encuesta

base=pd.read_csv('Universo_telcel_Nov2021.csv')
enc=pd.read_csv(ex_emp)
```

```python
# Ajustes y homologación de tablas

enc=enc.iloc[2:].reset_index(drop=True)
enc=enc[enc['EMP'].notna()]
enc['EMP']=enc['EMP'].astype('int')
base['EMP']=base['EMP'].astype('int')
```

```python
enc
```

```python
# Discriminación de encuestas realizadas

base_dep=base[~base['EMP'].isin(enc['EMP'])]
```

```python
base_dep
```

```python
len(base)-len(base_dep)
```

```python
# Inserta ruta y nombre del archivo de salida 'Muestra_Exp_mm_aa.csv' correspondiente al periodo.

muestra='Muestra_Exp_11_21.csv'
```

```python
# Selección y exportación de muestra aleatoria
a=base_dep.sample(2000)
```

```python
a['FECHA_NPS']='2021 11'
```

```python
# Selección y exportación de muestra aleatoria
a.to_csv(muestra,index=False)
```
