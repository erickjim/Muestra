# Proceso de Encuesta "Experiencia Empleado"

```python
import pandas as pd
```

```python
# Lectura de Universo Telcel y Encuesta

base=pd.read_csv('Universo_telcel_Nov2021.csv')
```

```python
m=base.sample(frac=.2)
muestras=[m]
for _ in range(3):
    m=base[~base['EMP'].isin(m['EMP'])].sample(4840)
    muestras.append(m)
    m=pd.concat(muestras)
muestras.append(base[~base['EMP'].isin(m['EMP'])])
```

```python
# Inserta ruta y nombre del archivo de salida 'Muestra_Exp_mm_aa.csv' correspondiente al periodo.
for i in range(5):
    muestra='Muestras/Muestra_Exp_'+f'0%i'%(i+1)+'_22.csv'
    a=muestras[i].copy()
    a['FECHA_NPS']='2022 '+f'0%i'%(i+1)
    a.to_csv(muestra,index=False)
```
